*Problem 4*

Explain Dimensionality Reduction. Perform PCA on the dataset and extract the proportion of variance explained by each principal component. How many principal components should be retained based on the `scree plot`? Examine the loadings of the first two principal components. Which variables contribute most to these components? (2 points)

**Hint**:
Scree Plot is a common method for determining the number of PCs to be retained through a graphical representation. A Scree Plot is a simple line segment plot that shows the eigenvalues for each individual PC. 

You can learn more about a scree plot [here](https://sanchitamangale12.medium.com/scree-plot-733ed72c8608)

**Hint**:
In PCA, the contribution of each feature to a principal component is called the `loading`. Loadings are compared with the absolute value



````r
# Your answer here
Attributes<- subset(Dataframes,select=-quality)
PCA_result <- prcomp(Attributes,scale=TRUE)
#screeplot(PCA_result,type="line",main="Scree Plot")
screeplot(PCA_result,main="Scree plot",type="line")
loadings <- PCA_result$rotation[, 1:2]
#print(loadings)
#Find the absolute value of loadings for PC1 &PC2
loadings_absolute<-abs(loadings)
#print(loadings_absolute)
sorted_PC1_loadings_abs <-sort(loadings_absolute[,1],decreasing=TRUE)
sorted_PC2_loadings_abs <-sort(loadings_absolute[,2],decreasing=TRUE)

#Creating a plot for better understanding
loading_df <- data.frame(
  Variable = names(sorted_PC1_loadings_abs),
  PC1 = sorted_PC1_loadings_abs,
  PC2 = sorted_PC2_loadings_abs[match(names(sorted_PC1_loadings_abs), names(sorted_PC2_loadings_abs))]
)

# Melting the data 
library(reshape2)
melted_loading_df <- melt(loading_df, id.vars = "Variable")

library(ggplot2)

# Bar plotting for both sorted PC1 and PC2 values
ggplot(melted_loading_df, aes(x = reorder(Variable, -value), y = value, fill = variable)) +
  geom_bar(stat = "identity", position = "dodge") +
  coord_flip() +
  labs(
    title = "Comparative Absolute Loadings for PC1 and PC2",
    x = "Variables",
    y = "Absolute Loading Value"
  ) +
  scale_fill_manual(values = c("PC1" = "blue", "PC2" = "red")) +
  theme_minimal()






````

Dimensionality reduction: When we have a vector of a higher dimension, we reduce the vector to a lower dimension not only for ease in calculations and mapping purpose .What dimensionality reduction does is it filters out the least contributing component(s) of a vector .Hence vector operations or manipulations,predictions   Based on the scree plot we can derive following conclusions
* The attributes ranging from 1 to 5 contribute most to the variance .
* The attribute 5 lies in the elbow region of the Scree plot
* Component 1 contributes most to the variance over component 2 as observable from bargraphs where total contribution of PC1 would be somewhere around 3.30 and of PC2 would be of 2.80
* Hence based on above observations we select components from 1-5.


*Problem 5* 

Now, for the selected components, find the percentage contribution of each feature to each of the components. (Hint: First find the square of the loadings also known as the cos2 values - find percentage contribution using these values) (2 points)

````r
# Your answer here# Your answer here
top_loadings <- PCA_result$rotation[,1:5]
#converting to cos2 values of loadings
cos2_values<- top_loadings^2
#calculating the percentage contribution by first calculating the totals
total_row_sums<- rowSums(cos2_values)

percentage_contribution<- (cos2_values/total_row_sums)*100
#print(percentage_contribution)
#Create a Dataframe for the contributions for easy understandings
contribution_dataframe<- as.data.frame(percentage_contribution)
contribution_dataframe$Variable<- rownames(contribution_dataframe)

library(reshape2)
melted_contribution_dataframe <- melt(contribution_dataframe,id.vars="Variable")
colnames(melted_contribution_dataframe)<-c("Variable", "PrincipalComponent", "PercentageContribution")
melted_contribution_dataframe


````
